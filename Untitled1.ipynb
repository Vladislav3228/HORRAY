```
from math import sqrt


def dist(a, b):
  D = 0
  for i in range(a.shape[0]):
    D += (a[i] - b[i])*(a[i] - b[i])
  return sqrt(D)

#p_1 - это искомая точка
def k(p_1, p_2, k_1):
  ch = dist(p_1, p_2)
  zn = dist(p_1, k_1)
  return 0.75*(1 - (ch/zn)*(ch/zn))

def recognize(k, p, data, target):
  q = data
  w = list() 
  for i in k:
    dis = 10000
    cur_j = 0;
    for j in q:
      new_dis = dist(p, data[j])
      if(new_dis < dis):
        dis = new_dis
        cur_j = j
    q.pop[cur_j]
    w.append(cur_j)
  
```
